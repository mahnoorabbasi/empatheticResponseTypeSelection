```
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
from scipy.sparse.csr import csr_matrix

# tfidf = TfidfVectorizer(sublinear_tf=True, min_df=5, norm='l2', encoding='latin-1', ngram_range=(1, 2), stop_words='english')

# features = tfidf.fit_transform(df.content).toarray()
vectorizer = TfidfVectorizer(analyzer='word', ngram_range=(1,1),
                     min_df = 0, stop_words='english')
tfidf_matrix =  vectorizer.fit_transform(df.content).toarray()
vocab = vectorizer.get_feature_names()
# print(vocab.sort)
```

```
# print(len(vocab))
# print(vocab.index('anger'))
# vocab[1322]
```

```
documents_tfidf_lol = [{word:tfidf_value for word, tfidf_value in zip(vocab, sent)} 
                       for sent in tfidf_matrix]

documents_tfidf = pd.DataFrame(documents_tfidf_lol)
documents_tfidf.fillna(0, inplace=True)

documents_tfidf
```

```
arousal_temp = documents_tfidf.copy()
for word in vocab:
  try:
    a_score_obj=a_scores.loc[word]
    a_score=a_score_obj[0]
  except KeyError:
    a_score = 0
  tfidfColScore=documents_tfidf[word]
  totalScore=a_score*tfidfColScore
  arousal_temp[word]=totalScore
  # finalDFrame=[word ,a_score ]
  print(word +" " +str(a_score))

```

```
arousal_temp["EmotionalValue"] = arousal_temp.sum(axis=1)
print(arousal_temp.sum(axis=1).idxmax())
arousal_temp.loc[0]
arousal_temp['EmotionalValue']
# temp['year']
```
